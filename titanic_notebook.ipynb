## Limpeza de Dados

Nesta etapa de **Limpeza de Dados**, realizamos uma série de transformações no conjunto de dados do Titanic com o objetivo de prepará-lo para a análise estatística e modelagem preditiva. Essa fase é essencial para garantir a **qualidade, consistência e representatividade** dos dados.

Primeiramente, identificamos e removemos **linhas duplicadas** e examinamos a presença de **valores ausentes** (NaN), tanto no conjunto de treino quanto no de teste.

Em seguida, realizamos as seguintes transformações e limpezas:

- **Codificação de variáveis categóricas**: convertendo o sexo (`Sex`) e o porto de embarque (`Embarked`) para valores numéricos;
- **Remoção de colunas irrelevantes**: como `Cabin`, `Ticket`, `PassengerId`, e `Name`, que não contribuem diretamente para o modelo ou possuem muitos valores ausentes;
- **Extração e padronização de títulos** a partir da coluna `Name` (ex: "Mr", "Miss", "Mrs", "Rare", etc.), com posterior mapeamento para valores numéricos;
- **Preenchimento dos valores ausentes de idade (`Age`)** com a **mediana por título** – estratégia baseada em grupos com características similares.

Além disso, criamos **novas variáveis derivadas**:
- `PclassSex`: combinação da classe social e sexo;
- `FamilySize`: total de familiares a bordo (irmãos + pais + o próprio passageiro).

Também foi realizado um **balanceamento da variável alvo (`Survived`)** usando **undersampling**, reduzindo a classe majoritária (não sobreviventes) para o tamanho da classe minoritária (sobreviventes), a fim de evitar viés no treinamento de modelos preditivos.

Posteriormente, identificamos e removemos **outliers na variável `Fare`**, excluindo passageiros com tarifas muito altas (acima de 300), que poderiam distorcer a análise estatística.

Por fim, garantimos que os **dados finais não possuam valores nulos**, especialmente no conjunto de teste ajustado, e exibimos uma **matriz de correlação** atualizada após a limpeza para verificar a relação entre as variáveis numéricas.

Essas ações tornam os dados mais **estruturados, limpos e informativos**, prontos para serem utilizados em modelagens preditivas como classificação, regressão e clustering.

#Normalização de Dados



Para evitar que variáveis com escalas diferentes dominem o comportamento dos modelos, aplicamos três técnicas distintas de normalização:

- **MinMaxScaler**: Redimensiona os dados para o intervalo [0, 1], sensível a outliers;
- **StandardScaler**: Aplica a padronização Z-score (média = 0, desvio padrão = 1);
- **RobustScaler**: Usa mediana e intervalo interquartílico (IQR), sendo mais robusto a outliers.

Cada técnica foi aplicada sobre os dados de treino (`X_train`), e seus efeitos foram visualizados por meio de **gráficos de boxplot**, permitindo comparar como cada abordagem lida com a dispersão e presença de valores extremos.

Essa preparação é essencial para que os algoritmos de machine learning tenham **dados bem comportados, equilibrados e comparáveis**, tanto durante o treinamento quanto nas etapas de validação e testes finais.

#Redes Neurais


Nesta seção, desenvolvemos e configuramos uma **Rede Neural do tipo MLP (Multilayer Perceptron)** utilizando o framework **Keras** com backend TensorFlow. O objetivo é construir um modelo capaz de prever a sobrevivência dos passageiros do Titanic com base nas variáveis numéricas previamente tratadas e normalizadas.

---

### 🔹 Arquitetura da Rede

A rede neural é composta por três camadas principais:

1. **Camada densa (Dense) com 32 neurônios**
   - Função de ativação: `ReLU`
   - Inicialização dos pesos: `HeNormal` (ideal para ReLU)
   - Regularização L2 (penaliza pesos grandes para evitar overfitting)

2. **Camada densa com 64 neurônios**
   - Função de ativação: `ReLU`
   - Inicialização dos pesos: `HeNormal`
   - Também com regularização L2

3. **Camada de saída com 1 neurônio**
   - Função de ativação: `sigmoid` (indicando probabilidade de sobrevivência)
   - Inicialização: distribuição normal (`RandomNormal`)




```python
import tensorflow as tf
import keras
from keras.layers import Dense, Input
from keras.regularizers import l2
from keras.models import Sequential
from tensorflow.keras import layers, callbacks, initializers
import random
import numpy as np
```

```python
tf.random.set_seed(SEED)
# Defini a arquitetura da MLP
model = Sequential([
    Dense(32, input_shape=(X_train_r.shape[1],),
          activation='relu',
          kernel_initializer=initializers.HeNormal(seed=SEED),
          kernel_regularizer=l2(0.01)),
    Dense(64, activation='relu',
          kernel_initializer=initializers.HeNormal(seed=SEED),
          kernel_regularizer=l2(0.01)),
    Dense(1, activation='sigmoid',kernel_initializer=initializers.RandomNormal(mean=0.0, stddev=0.05, seed=42))
])

model.compile(loss = tf.keras.losses.binary_crossentropy, optimizer = tf.keras.optimizers.Adam(learning_rate=0.001), metrics = ['acc'])

# model.save_weights('model_weights.weights.h5')
# model.load_weights('model_weights.weights.h5')

# Verificar pesos
for i, layer in enumerate(model.layers):
    weights = layer.get_weights()
    print(f"Layer {i} ({layer.name}):")
    if weights:
        print(f"  Weights shape: {weights[0].shape}")
        print(f"  Bias shape:    {weights[1].shape}")
        # Mostrar apenas alguns valores para não poluir a saída
        print(f"  First weight: {weights[0][0][0]:.5f}")
        print(f"  First bias:    {weights[1][0]:.5f}")
    else:
        print("  No weights")
```
